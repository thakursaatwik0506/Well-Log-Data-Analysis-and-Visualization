```python
#Project on Well Log Interpretation.
#The Project involves the use of various python libraries as welly, seaborn, numpy, pandas, matplotlib and warnings.
#The project works on the basis of taking in Well Log data in .LAS format and converting it first into a format readable by python.
#The code then works to define data sets in terms of Data frames and shows in heatmaps of the taken in input.
#Then the general signatures of the well log is plotted upon the conditions provided in the plot.
#This plot shown has sample data taken from the input and output computation of the Volve 15/9-F-1B well.
```

```python
import welly
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
```

```python
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#Taking in well data in LAS format
welli= welly.Well.from_las('Sample Log Input Data.LAS')
```

```python
dfi= welli.df()
```

```python
dfi
```

```python
type(welli)
```

```python
welli
```

```python
#Plotting the heatmap
sns.heatmap(dfi.isna())
```

```python
welli.count_curves()
```

```python
#Plotting well log data
pl.style.use('Solarize_Light2')
welli.plot()
```

```python
#Plotting signature tracks
tracks= ['MD','BS','CALI',['GR','RHOB'],'NPHI','RD','ROP']
welli.plot(tracks=tracks)
```
